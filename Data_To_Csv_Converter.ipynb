```python
import cv2
import os
import numpy as np
import pandas as pd
```

```python
brain_tumor_train_path = "./Brain scans\Tumor\TRAIN"
brain_tumor_test_path = "./Brain scans\Tumor\TEST"
brain_notumor_train_path = "./Brain scans/No tumor\Train"
brain_notumor_test_path = "./Brain scans/No tumor\Test"
breast_benign_train_path = "./Breast scans\\benign\Train"
breast_benign_test_path = "./Breast scans\\benign\Test"
breast_malignant_train_path = "./Breast scans\malignant\Train"
breast_malignant_test_path = "./Breast scans\malignant\Test"
breast_normal_train_path = "./Breast scans/normal\Train"
breast_normal_test_path = "./Breast scans/normal\Test"
```

```python
train_data = []
test_data = []
```

```python
#appending training data
for filename in os.listdir(brain_tumor_train_path):
    train_data.append([brain_tumor_train_path +'\\'+ filename,"brain" ,"tumor"])
for filename in os.listdir(brain_notumor_train_path):
    train_data.append([brain_notumor_train_path +"\\"+ filename,"brain", "no tumor"])
for filename in os.listdir(breast_benign_train_path):
    train_data.append([breast_benign_train_path +'\\'+ filename,"breast" ,"benign"])
for filename in os.listdir(breast_malignant_train_path):
    train_data.append([breast_malignant_train_path +"\\"+ filename,"breast", "malignant"])
for filename in os.listdir(breast_normal_train_path):
    train_data.append([breast_normal_train_path +'\\'+ filename,"breast" ,"normal"])
    
#appending test data  
for filename in os.listdir(brain_tumor_test_path):
    test_data.append([brain_tumor_test_path +"\\"+ filename,"brain", "tumor"])
for filename in os.listdir(brain_notumor_test_path):
    test_data.append([brain_notumor_test_path +"\\"+ filename,"brain", "no tumor"])
for filename in os.listdir(breast_benign_test_path):
    test_data.append([breast_benign_test_path +'\\'+ filename,"breast" ,"benign"])
for filename in os.listdir(breast_malignant_test_path):
    test_data.append([breast_malignant_test_path +"\\"+ filename,"breast", "malignant"])
for filename in os.listdir(breast_normal_test_path):
    test_data.append([breast_normal_test_path +'\\'+ filename,"breast" ,"normal"])
```

```python
#generating dataframes
train_df = pd.DataFrame(train_data, columns=["path","type", "diagnose"])
test_df = pd.DataFrame(test_data, columns=["path","type", "diagnose"])
train_df
```

```python
train_df.to_csv('train.csv', index = False)
test_df.to_csv('test.csv', index = False)
```
